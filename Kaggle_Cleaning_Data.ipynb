**Why we have missing values???**

We can have a missing value for two reason mainly:
 

*   Because it doesn' exist (like a child for someone that doesn't have a child)
*   Because wasn't recorded and we need to calculate it from other values in the column

In the next lines of code I verify some columns to verify which kind of missing values we have



```
# I verify column "Street Number Suffix". From the previous analysis, missing values: 196684/198900
# We try to visualize not null values 
j = 0
i = 0
for _ in sf_permits[sf_permits.columns[7]].isnull():
  if _ == False:
    print(sf_permits[sf_permits.columns[7]][i])
    j = j+1
  i = i+1
  if j == 10:
    break
    
```

```
#Print first 30 values of "Street Number Suffix" with no NULL values and more columns to understand the meaning

j = 0
i = 0
for _ in sf_permits[sf_permits.columns[7]].isnull():
  if _ == False:
    print(sf_permits[sf_permits.columns[6]][i],sf_permits[sf_permits.columns[7]][i],sf_permits[sf_permits.columns[8]][i],sf_permits[sf_permits.columns[9]][i],sf_permits[sf_permits.columns[10]][i])
    j = j+1
  i = i+1
  if j == 30:
    break

```

```
#Print 20 values of "Street Number Suffix" with NULL values and more columns to understand if they doesn't exist or aren't recorded
#The values are missing values because they aren't recorded!!!
j = 0
i = 0
for _ in sf_permits[sf_permits.columns[7]].isnull():
  if _ == True:
    print(sf_permits[sf_permits.columns[6]][i],sf_permits[sf_permits.columns[7]][i],sf_permits[sf_permits.columns[8]][i],sf_permits[sf_permits.columns[9]][i],sf_permits[sf_permits.columns[10]][i])
    j = j+1
  i = i+1
  if j == 20:
    break
```

```
#Now we check Zipcode, print number of null value and percentageof NULL values on total values
#We have a low percentage of NULL value
total_cells = np.product(sf_permits.shape[0])
total_missing = (sf_permits[sf_permits.columns[40]].isnull().sum()/total_cells)*100
print("Number of missing values:",sf_permits[sf_permits.columns[40]].isnull().sum(),"/",total_cells)
print("Percentage of missing values for Zipcode column:",total_missing)
```

```
#Print 20 values of "ZipCode" with NULL values and more columns (Street Name) to understand if they doesn't exist or aren't recorded
#ZipcCode is NUL because these values aren't recorded!!!
j = 0
i = 0
for _ in sf_permits[sf_permits.columns[40]].isnull():
  if _ == True:
    print(sf_permits[sf_permits.columns[40]][i],sf_permits[sf_permits.columns[8]][i])
    j = j+1
  i = i+1
  if j == 20:
    break
```
